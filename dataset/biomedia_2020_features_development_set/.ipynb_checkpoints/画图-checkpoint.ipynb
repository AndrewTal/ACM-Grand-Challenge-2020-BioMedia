```python
def data_convert(x):
    try:
        result = float(str(x).replace(',', '.')) if isinstance(x, np.str) else x
    except:
        result = 0
    return result
```

```python
def corr_plot(x_name, y_name, data):
    plt.figure(dpi=500)
    JG = sns.jointplot(x = x_name, 
                  y = y_name,
                  data = data, 
                  stat_func=sci.pearsonr, 
                  kind = 'reg')
    JG.savefig(x_name + '.png')
```

```python
semen_analysis_data = pd.read_csv('./semen_analysis_data.csv', sep = ';').applymap(data_convert)
serum_data = pd.read_csv('./fatty_acids_serum.csv', sep = ';').applymap(data_convert)

data = pd.concat([semen_analysis_data, 
                  serum_data], axis=1).T.drop_duplicates().T
```
